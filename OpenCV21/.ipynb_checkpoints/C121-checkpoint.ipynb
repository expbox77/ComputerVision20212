# 1.2.1 Getting Started with Images
### cv2.imread(), cv2.imshow() , cv2.imwrite()

```python
# Read an image
```

```python
import numpy as np
import cv2
```

```python
# Load an color image in grayscale
# 플래그 -1 -> UNCHANGED
# 플래그 0 -> GRAYSCALE
# 플래그 1 -> COLOR
img = cv2.imread('cat.jpg', 0)
cv2.imshow('image', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.waitKey(1)
```

```python
# cv2.WINDOW_NORMAL
cv2.namedWindow('image', cv2.WINDOW_NORMAL)
cv2.imshow('image', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.waitKey(1)
```

```python
# Write an image
```

```python
# cv2.imwrite('catgray.jpg', img)
```

```python
img = cv2.imread('cat.jpg', 0)
cv2.imshow('image', img)
k = cv2.waitKey(0)
```

```python
# ESC를 눌러 종료
if k == 27:
    cv2.destroyAllWindows()
    cv2.waitKey(1)

# s키룰 눌러 저장 후 종료
elif k == ord('s'):
    cv2.imwrite('catgray.png', img)
    cv2.destroyAllWindows()
    cv2.waitKey(1)
```

```python
# Using Matplotlib
```

```python
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('cat.jpg', 0)
plt.imshow(img, cmap = 'gray', interpolation = 'bicubic')

# to hide tick values on X and Y axis
plt.xticks([]), plt.yticks([])
plt.show()
```
