# 1.4.7 Canny Edge Detection
### cv2.Canny()

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('D:\GitHub\ComputerVision20212\SampleFiles\cat.jpg',0)
edges = cv2.Canny(img,200,300)  # min --> 200 / max --> 300

plt.rcParams["figure.figsize"] = (15, 10)

plt.subplot(121),plt.imshow(img,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(edges,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('D:\GitHub\ComputerVision20212\SampleFiles\cat.jpg',0)

n = 3
gaussian = cv2.GaussianBlur(img, (n, n), 0)

edges = cv2.Canny(gaussian,200,300)  # min --> 200 / max --> 300

plt.rcParams["figure.figsize"] = (15, 10)

plt.subplot(121),plt.imshow(img,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(edges,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])

plt.show()
```
