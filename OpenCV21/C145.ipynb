# 1.4.5 Morphological Transformations
### cv2.erode(), cv2.dilate(), cv2.morphologyEx() etc.

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('j.png',0)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Erosion --> 객체 외곽을 깎아내는 후처리 연산
# 노이즈(잡음)를 제거하거나 붙어있는 오브젝트들을 분리하는데 효과가 있음.

n = 3
kernel = np.ones((n,n),np.uint8)

# cv2.erode(이미지, 커널 크기, iterations=(반복횟수))
erosion = cv2.erode(img,kernel,iterations = 1)
erosion2 = cv2.erode(img,kernel,iterations = 2)

plt.rcParams["figure.figsize"] = (15, 10)

plt.subplot(131),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])

plt.subplot(132),plt.imshow(erosion),plt.title('Erosion')
plt.xticks([]), plt.yticks([])

plt.subplot(133),plt.imshow(erosion2),plt.title('Erosion2')
plt.xticks([]), plt.yticks([])

plt.show()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('j.png',0)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Dilation --> 객체 외각을 확대시키는 후처리 연산
# Erosion에 의해서 작아졌던 오브젝트를 원래대로 돌리거나 인접해 있는 오브젝트들을 하나로 만드는데 사용 가능.

n = 3
kernel = np.ones((n,n),np.uint8)

# cv2.dilate(이미지, 커널, iterations=(반복횟수))
dilation = cv2.dilate(img,kernel,iterations = 1)
dilation2 = cv2.dilate(img,kernel,iterations = 2)

plt.rcParams["figure.figsize"] = (15, 10)

plt.subplot(131),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])

plt.subplot(132),plt.imshow(dilation),plt.title('Dilation')
plt.xticks([]), plt.yticks([])

plt.subplot(133),plt.imshow(dilation),plt.title('Dilation2')
plt.xticks([]), plt.yticks([])

plt.show()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('j.png',0)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Opening --> Erosion 연산 다음 Dilation 연산을 적용하여 이미지 상의 노이즈를 제거하는데 사용한다.

n = 5
kernel = np.ones((n,n),np.uint8)

# cv2.morphologyEx(이미지, 모폴로지 연산 종류, 커널)
# Opening --> cv2.MORPH_OPEN
opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)

plt.rcParams["figure.figsize"] = (10, 10)

plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])

plt.subplot(122),plt.imshow(opening),plt.title('Opening')
plt.xticks([]), plt.yticks([])

plt.show()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('j.png',0)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Closing --> Dilation 연산 다음 Erosion 연산을 적용하여 이미지 상에 있는 작은 검은색 구멍들을 메우는데 사용된다.

n = 5
kernel = np.ones((n,n),np.uint8)

# cv2.morphologyEx(이미지, 모폴로지 연산 종류, 커널)
# Closing --> cv2.MORPH_CLOSE

closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)

plt.rcParams["figure.figsize"] = (10, 10)

plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])

plt.subplot(122),plt.imshow(closing),plt.title('Closing')
plt.xticks([]), plt.yticks([])

plt.show()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('j.png',0)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# 팽창 연산을 적용한 이미지에서 침식 연산을 적용한 이미지를 빼어 경계를 검출한다.
# 이런 연산을 그레디언트 연산이라고 한다.

n = 5
kernel = np.ones((n,n),np.uint8)

# cv2.morphologyEx(이미지, 모폴로지 연산 종류, 커널)
# Gradient --> cv2.MORPH_GRADIENT

gradient = cv2.morphologyEx(img, cv2.MORPH_GRADIENT, kernel)

plt.rcParams["figure.figsize"] = (10, 10)

plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])

plt.subplot(122),plt.imshow(gradient),plt.title('Gradient')
plt.xticks([]), plt.yticks([])

plt.show()
```
