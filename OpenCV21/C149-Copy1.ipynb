# 1.4.9 Contours in OpenCV
### cv2.findContours(), cv2.drawContours()

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt

img = cv2.imread('D:\GitHub\ComputerVision20212\SampleFiles\logo.png')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
imgray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
imgray = 255-imgray

plt.figure(figsize = (15,15))
plt.subplot(1,2,1),plt.imshow(img)
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.subplot(1,2,2),plt.imshow(imgray,cmap = 'gray')
plt.title('Gray'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
ret,thresh = cv2.threshold(imgray,8,255,cv2.THRESH_BINARY)
contours, hierachy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
```

```python
imgcontour = cv2.drawContours(img,contours,-1,(255,32,32),4)
```
