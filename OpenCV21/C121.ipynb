# 1.2.1 Getting Started with Images
### cv2.imread(), cv2.imshow() , cv2.imwrite()

##### Using OpenCV

```python
# Read an image
```

```python
import numpy as np
import cv2
```

```python
# Load an color image in grayscale
# cv2.imread('파일명', 플래그)
# 플래그 -1 -> UNCHANGED
# 플래그 0 -> GRAYSCALE
# 플래그 1 -> COLOR

img = cv2.imread('cat.jpg', 0)
```

```python
# Display an image
```

```python
# cv2.imshow('띄울 창의 이름', 이미지가 저장된 변수)

cv2.imshow('image', img)

# cv2.waitKey(X 밀리초 대기) 키보드 입력을 기다림
# cv2.waitKey(0)일 때는 무한 대기

cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.waitKey(1)
```

```python
# cv2.WINDOW_NORMAL: 사용자가 창 크기를 조정할 수 있음
# cv2.WINDOW_AUTOSIZE: 이미지와 동일한 크기로 창 크기를 재조정할 수 없음

cv2.namedWindow('image', cv2.WINDOW_NORMAL)
cv2.imshow('image', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.waitKey(1)
```

```python
# Write an image
```

```python
# cv2.imwrite('출력할 파일명', 이미지를 불러온 변수)

cv2.imwrite('catgray1.png', img)
```

```python
# Sum it up
```

```python
import numpy as np
import cv2

# img 변수에 cat.jpg를 플래그 0(GrayScale)로 불러오기
img = cv2.imread('cat.jpg', 0)

# img 변수에 있는 cat.jpg를 image라는 창으로 띄우기
cv2.imshow('image', img)

k = cv2.waitKey(0)

# ESC(27)를 눌러 종료
if k == 27:
    cv2.destroyAllWindows()
    cv2.waitKey(1)

# s키를 눌러 저장 후 종료
# ord('문자') 함수 --> 특정한 한 문자를 아스키 코드 값으로 변환해 주는 함수

elif k == ord('s'):
    cv2.imwrite('catgray2.png', img)
    cv2.destroyAllWindows()
    cv2.waitKey(1)
```

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt

img = cv2.imread('cat.jpg', 0)
plt.imshow(img, cmap = 'gray', interpolation = 'bicubic')

# X축과 Y축의 눈금 값을 숨기기
plt.xticks([]), plt.yticks([])
plt.show()
```
