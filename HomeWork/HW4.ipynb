### Remove Perspective Distortion

```python
# import modules
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Read Image
img = cv2.imread('SampleFiles/HW4/dsu4.jpg')
```

```python
# pts1 --> 왜곡된 이미지의 각 꼭짓점 죄표
pts1 = np.float32([[330,260],[570,230],[565,495], [330,460]])

# pts2 --> 왜곡된 이미지를 폈을 때 꼭짓점 좌표
pts2 = np.float32([[0,0],[500,0],[500,300],[0,300]])
```

```python
# Perspective Remove
M = cv2.getPerspectiveTransform(pts1,pts2)
dst = cv2.warpPerspective(img,M,(500,300))
```

```python
# Output as matplotlib.
plt.rcParams["figure.figsize"] = (20,10)
plt.subplot(121),plt.imshow(img),plt.title('INPUT')
plt.subplot(122),plt.imshow(dst),plt.title('OUTPUT')

plt.show()
```
